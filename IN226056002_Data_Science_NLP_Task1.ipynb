## 1. In NLP, "Love" and "love" are treated as different words because they are case-sensitive. "Love"(Uppercase L) - may appear at the beginning of a sentence or as a proper noun. "love"(lowercas l ) - used as a regular word verb or noun. Case differences can create duplicate features and affect model accuracy,so it is converted to lower case.

## 2. Stopwords are extremely common words like "the", "is", "and", "a", "in" etc. If we keep them in the data: They will show up at the top of every frequency count and dominate the results, They push actual meaningful words down in importance, Models trained on this data end up learning that "the" and "is" are super important features, which is obviously wrong, It also bloats the vocabulary size unnecessarily which slows everything down. Signal to noise ratio drops a lot when stopwords are kept.

## 3. Sentiment Analysis: "The service was not good." If you strip the stopwords, "not" is often deleted. The model is left with "service good," which is the exact opposite of what the user meant. Search Engines & Quotes: If someone searches for the famous line "To be or not to be," and system removes stopwords, the search query becomes empty. Here, the filler words are the message.

## 4. Stemming just chops off the end of the word using fixed rules — it doesnt care if the result is a real word or not. So "studies" becomes "studi" and "running" becomes "run" but "fairly" might become "fairli" which is an invalid word. Lemmatization looks the word up in a dictionary and returns its proper base form called the lemma. So "studies" becomes "study", "running" becomes "run", "better" becomes "good". The output is always a real word.

```python
def get_analytics(tokens):
    if not tokens:
        return 0, 0, 0.0
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total_tokens
    return total_tokens, unique_tokens, round(avg_length, 2)
def preprocess_and_analyze(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    exceptions = {'no', 'not'}
    tokens = [w for w in text.split() if len(w) > 2 or w in exceptions]
    total, unique, avg_len = get_analytics(tokens)
    return {
        "tokens": tokens,
        "total": total,
        "unique": unique,
        "avg_len": avg_len
    }
print(f"{'Original Sentence':<35} | {'Tokens':<5} | {'Unique':<6} | {'Avg Len'}")
print("-" * 70)
for sentence in test_cases:
    data = preprocess_and_analyze(sentence)
    print(f"{sentence[:34]:<35} | {data['total']:<5} | {data['unique']:<6} | {data['avg_len']}")
```

```python
from collections import Counter
def clean_and_tokenize(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    exceptions = {'no', 'not'}
    tokens = [w for w in text.split() if len(w) > 2 or w in exceptions]
    return tokens
all_cleaned_tokens = []
for sentence in test_cases:
    tokens = clean_and_tokenize(sentence)
    all_cleaned_tokens.extend(tokens)
word_counts = Counter(all_cleaned_tokens)
top_10 = word_counts.most_common(10)
least_5 = sorted(word_counts.items(), key=lambda x: x[1])[:5]
print("--- TOP 10 MOST FREQUENT WORDS ---")
for word, count in top_10:
    print(f"{word:<12} | {count} times")
print("\n--- TOP 5 LEAST FREQUENT WORDS ---")
for word, count in least_5:
    print(f"{word:<12} | {count} time")
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    exceptions = {'no', 'not'}
    for text in text_list:
        text = text.lower()
        text = re.sub(r'https?://\S+|www\.\S+', '', text)
        text = re.sub(r'\S+@\S+\.\S+', '', text)
        text = re.sub(r'(.)\1{2,}', r'\1', text)
        text = re.sub(r'\d+', '', text)
        text = re.sub(r'[^\w\s]', '', text)
        current_tokens = [
            word for word in text.split()
            if len(word) > 2 or word in exceptions
        ]
        clean_sentence = " ".join(current_tokens).strip()
        all_tokens.extend(current_tokens)
        if clean_sentence:
            clean_sentences.append(clean_sentence)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
output = full_pipeline(test_cases)
print("--- CLEANED SENTENCES ---")
for i, s in enumerate(output['clean_sentences'], 1):
    print(f"{i}. {s}")
print("\n--- TOKEN PREVIEW (First 15) ---")
print(output['tokens'][:15])
```

```python
def improved_pipeline(text_list):
    if not isinstance(text_list, list):
        return {"error": "Input must be a list of strings."}
    all_tokens = []
    clean_sentences = []
    exceptions = {'no', 'not'}
    for text in text_list:
        if not text or not str(text).strip():
            continue
        text = str(text).lower()
        text = re.sub(r'https?://\S+|www\.\S+', '', text)
        text = re.sub(r'\S+@\S+\.\S+', '', text)
        text = re.sub(r'(.)\1{2,}', r'\1', text)
        text = re.sub(r'\d+', '', text)
        text = re.sub(r'[^\w\s]', '', text)
        current_tokens = [
            word for word in text.split()
            if len(word) > 2 or word in exceptions
        ]
        if not current_tokens:
            continue
        clean_sentence = " ".join(current_tokens).strip()
        all_tokens.extend(current_tokens)
        clean_sentences.append(clean_sentence)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences,
        "processed_count": len(clean_sentences)
    }
edge_cases = [
    "",
    "😍😍😍",
    "123456",
    None,
    "",
    "   ",
    "🚀🚀🚀😍😍",
    "1234567890",
    "!!! ??? @#$",
    "I am not happy"
]
result = improved_pipeline(edge_cases)
print(f"Total Sentences Input: {len(edge_cases)}")
print(f"Total Sentences Successfully Processed: {result['processed_count']}")
print(f"Final Clean Sentences: {result['clean_sentences']}")
```
