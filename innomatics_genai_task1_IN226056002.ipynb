```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

```python
import os
os.environ["GROQ_API_KEY"] = ""
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

chain1 = explain_template | model

response = chain1.invoke({"topic": "Neural Networks"})
print(response.content)
```

```python
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

chain2 = multi_input_template | model

test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

for case in test_cases:
    print(f"\nInput: {case}")
    response = chain2.invoke(case)
    print(response.content)
```

```python
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

for name, template in {
    "Teaching": teaching_template,
    "Interview": interview_template,
    "Storytelling": storytelling_template
}.items():
    chain = template | model
    response = chain.invoke({"topic": topic})
    print(f"\n{name} Output:\n")
    print(response.content)
```

```python
role_map = {
    "teacher": "You are a helpful teacher who explains step by step.",
    "interviewer": "You are an interviewer who asks sharp and relevant questions.",
    "motivator": "You are a motivating coach who explains ideas with encouragement."
}

chat_template = ChatPromptTemplate.from_messages([
    ("system", "{system_role}"),
    ("user", "Explain {topic}")
])

role = "teacher"
topic = "Neural Networks"

chat_chain = chat_template | model

response = chat_chain.invoke({
    "system_role": role_map[role],
    "topic": topic
})

print(response.content)
```

```python
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    if audience not in valid_audiences:
        audience = "beginner"

    if tone not in valid_tones:
        tone = "formal"

    return audience, tone

print(validate_inputs("beginner", "fun"))
print(validate_inputs("kids", "friendly"))
```

```python
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Ask 3 interview questions about {topic} for {audience} in a {tone} style"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    )
}

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    if style not in style_templates:
        raise ValueError("Style must be one of: teaching, interview, storytelling")

    selected_template = style_templates[style]
    chain = selected_template | model
    response = chain.invoke({
        "topic": topic,
        "audience": audience,
        "tone": tone
    })
    return response.content

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
one_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

reuse_chain = one_template | model

inputs = [
    {"topic": "AI", "audience": "beginner", "tone": "formal"},
    {"topic": "Python", "audience": "intermediate", "tone": "casual"},
    {"topic": "Deep Learning", "audience": "expert", "tone": "formal"},
    {"topic": "Data Science", "audience": "beginner", "tone": "fun"},
    {"topic": "NLP", "audience": "expert", "tone": "casual"}
]

for item in inputs:
    print(f"\nInput: {item}")
    response = reuse_chain.invoke(item)
    print(response.content)
```
